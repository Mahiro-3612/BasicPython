```python
import seaborn as sns

plt.figure(figsize=(10, 6))

sns.histplot(hist_df["aデータ"], color='orange', alpha=0.3, binwidth=1.5, label='aデータ', edgecolor='none')

sns.histplot(hist_df["bデータ"], color='green', alpha=0.3, binwidth=1.5, label='bデータ', edgecolor='none')

plt.title('aデータとbデータの温度に関する度数分布', fontsize=16)
plt.xlabel('温度(°C)', fontsize=12)
plt.ylabel('度数', fontsize=12)

plt.legend()

plt.show()
```

## 2問目

```python
scatter_df = pd.read_csv("data/scatter_data.csv")
scatter_df.head(5)
```

```python
scatter_df = pd.read_csv("data/scatter_data.csv")

color_map = {0: 'red', 1: 'blue', 2: 'green'}
label_map = {0: 'class0', 1: 'class1', 2: 'class2'}

plt.figure(figsize=(10, 6))

for target in [0, 1, 2]:
    subset = scatter_df[scatter_df['class'] == target]
    sns.scatterplot(
        data=subset,
        x='length (cm)',
        y='width (cm)',
        color=color_map[target],
        marker="o",
        label=label_map[target],
        s=100  
    )

plt.title('クラス０〜２のlengthとwidthに関する散布図', fontsize=16)
plt.xlabel('length (cm)', fontsize=12)
plt.ylabel('width (cm)', fontsize=12)

plt.legend(title='Species', title_fontsize='13', fontsize='11')

plt.grid(True, linestyle='--')

plt.show()
```

```python
linedata_df = pd.read_csv("data/line_data.csv")
linedata_df.head(5)
```

```python
vivid_blue = (0/255, 150/255, 255/255) 

plt.figure(figsize=(10, 6))
sns.lineplot(x=linedata_df["期間"], y=linedata_df["件数"], color=vivid_blue, linewidth=2, markersize=10, marker='o')

plt.title('期間ごとの件数の推移', fontsize=16)
plt.xlabel('期間', fontsize=12)
plt.ylabel('件数', fontsize=12)
plt.grid(True, linestyle='-')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x=linedata_df["期間"], y=linedata_df["費用"], color=vivid_blue, edgecolor='none')

plt.title('期間ごとの費用の推移', fontsize=16)
plt.xlabel('期間', fontsize=12)
plt.ylabel('費用(円)', fontsize=12)
plt.grid(True, linestyle='-')
plt.show()
```
